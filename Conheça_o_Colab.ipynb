```
from google import genai
from google.colab import userdata

client = genai.Client(api_key=userdata.get('GOOGLE_API_KEY'))
```

```
from google.genai import types
instrucoes_sistema = """
Você é o 'Especialista em Garantia Ar-Condicionado'.
Seu conhecimento é restrito EXCLUSIVAMENTE às seguintes regras:
1. Garantia de 1 ano: Se o aparelho for instalado por um técnico NÃO credenciado.
2. Garantia de 2 anos: Se o aparelho for instalado por um técnico credenciado.
3. Requisito Obrigatório: É indispensável ter a nota fiscal em mãos para qualquer processo.

REGRAS DE COMPORTAMENTO:
- NÃO responda sobre preços, marcas específicas, instalação técnica ou outros assuntos.
- Se o usuário perguntar algo fora do tema, diga: "Sou especialista apenas em regras de garantia e nota fiscal desta marca."
- Você deve responder exatamente 3 perguntas.
- Na 3ª resposta, você DEVE obrigatoriamente:
  1. Dar a resposta técnica solicitada.
  2. Fazer um resumo das 3 dúvidas tratadas.
  3. Encerrar com a frase: [ATENDIMENTO FINALIZADO].
"""

config_bot = types.GenerateContentConfig(
    system_instruction=instrucoes_sistema,
    temperature=0.1,
    top_p=0.95,
    top_k=10,
)
```

```
chat = client.chats.create(model='gemini-2.5-flash', config=config_bot)

print("RR CLIMATIZAÇÃO Assistente de garantia: Olá! Sou seu assistente de garantia. Como posso ajudar?\n")

for i in range(3):
    pergunta_usuario = input(f"Pergunta {i+1}/3: ")
    response = chat.send_message(pergunta_usuario)
    print(f"\n RR CLIMATIZAÇÃO Assistente de Garantia: {response.text}\n")

print("-" * 40)
finalizacao = chat.send_message("Agora, faça o resumo das 3 perguntas anteriores e encerre a conversa conforme as instruções.")
print(f"RESUMO FINAL:\n{finalizacao.text}")
```
